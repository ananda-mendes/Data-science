```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Futebol e Dados/matches.csv')
df
```

```python
df.isnull().sum()
```

```python
df.drop(['match_time', 'tournament_id', 'match_date', 'stadium_name', 'country_name', 'country_id_iso_3', 'team_a_code', 'team_b_code', 'team_a_id_iso_3', 'team_b_id_iso_3', 'team_a_score_margin', 'team_b_score_margin'], axis=1, inplace=True)
df
```

```python
home_win = df.groupby('team_a_name')['team_a_win'].sum()
away_win = df.groupby('team_b_name')['team_b_win'].sum()

# Adicionando a Russia manualmente pois ela nunca jogou como mandante
home_win['Russia'] = 0

# Vitórias por país
country_win = home_win + away_win
country_win 
```

```python
import matplotlib.pyplot as plt

# Ordenar os países por número de vitórias
country_win_sorted = country_win.sort_values(ascending=False)

# Criar o gráfico de barras horizontais
plt.figure(figsize=(10, 8))
plt.barh(country_win_sorted.index, country_win_sorted.values, color='blue')
plt.xlabel('Número de Vitórias')
plt.ylabel('Países')
plt.title('Vitórias por País na história das Copas')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular o total de gols marcados por cada equipe
team_goals = df.groupby('team_a_name')['team_a_score'].sum().add(df.groupby('team_b_name')['team_b_score'].sum(), fill_value=0)

# Ordenar as equipes pelo total de gols marcados
team_goals = team_goals.sort_values(ascending=False).reset_index()
team_goals.columns = ['Equipe', 'Total de Gols Marcados']

plt.figure(figsize=(15, 5))
sns.barplot(x="Equipe", y="Total de Gols Marcados", data=team_goals)
plt.title("Seleções com maior número de gols marcados", color='black')
plt.xlabel('Equipe')
plt.ylabel('Total de Gols Marcados')
plt.xticks(rotation=45)
plt.show()
```

```python
games_by_year = df.groupby('year').size()

plt.figure(figsize=(12, 6))
plt.plot(games_by_year.index, games_by_year.values, marker='o', linewidth=2)
plt.xlabel('Ano')
plt.ylabel('Número de Jogos')
plt.title('Evolução do Número de Jogos ao longo dos Anos')
plt.xticks(rotation=45)
plt.show()

```

```python
# Filtrar os jogos da final
finals = df[df['stage_name'] == 'Final']

# Criar uma coluna para armazenar o time vencedor da final
df['champion'] = ''

# Iterar sobre os jogos da final
for index, row in finals.iterrows():
    if row['team_a_win'] == 1:
        df.loc[index, 'champion'] = row['team_a_name']
    elif row['team_b_win'] == 1:
        df.loc[index, 'champion'] = row['team_b_name']

# Contar o número de vezes que cada equipe venceu o torneio, excluindo o contador
champion_counts = df[df['champion'] != '']['champion'].value_counts()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x=champion_counts.index, y=champion_counts.values)
plt.title("Maiores Campeãs da Copa do Mundo Feminina")
plt.xlabel("Equipe")
plt.ylabel("Número de Títulos")
plt.xticks(rotation=45)
plt.show()
```

```python
# Filtrar os jogos da final
finals = df[df['stage_name'] == 'Final']

# Criar uma coluna para armazenar o time vice-campeão da final
df['runner_up'] = ''

# Iterar sobre os jogos da final
for index, row in finals.iterrows():
    if row['team_a_win'] == 0 and row['team_b_win'] == 1:
        df.loc[index, 'runner_up'] = row['team_a_name']
    elif row['team_a_win'] == 1 and row['team_b_win'] == 0:
        df.loc[index, 'runner_up'] = row['team_b_name']

# Contar o número de vezes que cada equipe foi vice do torneio, excluindo o contador
runner_up_counts = df[df['runner_up'] != '']['runner_up'].value_counts()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x=runner_up_counts.index, y=runner_up_counts.values)
plt.title("Maiores Vice-Campeãs da Copa do Mundo Feminina")
plt.xlabel("Equipe")
plt.ylabel("Número de Vice-Campeonatos")
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar os jogos da disputa de terceiro lugar
third_place = df[df['stage_name'] == 'Match for third place']

# Criar uma coluna para armazenar o terceiro lugar
df['third_place'] = ''

# Iterar sobre os jogos da disputa de terceiro lugar
for index, row in third_place.iterrows():
    if row['team_a_win'] == 1:
        df.loc[index, 'third_place'] = row['team_a_name']
    elif row['team_b_win'] == 1:
        df.loc[index, 'third_place'] = row['team_b_name']
    else:
        # Caso de empate, considerar os pênaltis
        penalties = row['score_penalties']
        if penalties.startswith(row['team_a_name']):
            df.loc[index, 'third_place'] = row['team_a_name']
        else:
            df.loc[index, 'third_place'] = row['team_b_name']

# Contar o número de vezes que cada equipe foi terceiro lugar do torneio, excluindo o contador
third_place_counts = df[df['third_place'] != '']['third_place'].value_counts() 

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x=third_place_counts.index, y=third_place_counts.values)
plt.title("Equipes com Mais Terceiros Lugares na Copa do Mundo Feminina")
plt.xlabel("Equipe")
plt.ylabel("Número de Terceiros Lugares")
plt.xticks(rotation=45)
plt.show()
```

```python
# Aqui, vamos concatenar as três colunas que estrutramos acima nos gráficos.
# Além disso, transformaremos todos os campos NAN's em 0 e transformaremos todos os valores em inteiros
teams = pd.concat([champion_counts, runner_up_counts, third_place_counts], axis=1)
teams.fillna(0, inplace=True)
teams = teams.astype(int)
teams
```
